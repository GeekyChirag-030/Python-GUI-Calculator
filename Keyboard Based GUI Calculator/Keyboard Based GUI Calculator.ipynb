```python
Window=Tk()

D=StringVar()
Window.geometry('600x600+400+100') #Setting Size of Window With the Position of Initial Window

#Set the Icon on Main Parent Window 
Window.iconbitmap("C:\\Users\HP\\Python Additional Projects (For Practice)\\Python Calculator Project\\Calculator_1.ico")

#Set Title and Background Colour of Tkinter Window

Window.title("            --------------------   🎰    Scientific Calculator   💰   -------------------        ")
Window['bg']="moccasin"

Label(Window,text="   Scientific Keyboard Based Calculator   ",font=("Times New Roman",40,"bold","italic"),cursor="bogosity",bd=9,bg="white",fg="black",relief="groove",highlightthickness=5.5,highlightcolor="black").pack(padx=10,pady=20)

#Main Code of Calc
Label(Window,text=" **   Enter Your Equation or Statement Whose Mathematical Computation You Want To Perform ** ",font=("Typiko New Era",23,"bold"),cursor="dot",fg="black",bg="beige",bd=7.5,highlightthickness=3,highlightcolor="black",relief="groove").pack(padx=40,pady=30)

Enter=Entry(Window,font=("Arial",33,"bold","italic"),textvariable=D,cursor="xterm",relief="groove",bd=10,selectbackground="blue",highlightthickness=8,highlightcolor="darkslategray",selectforeground="yellow",selectborderwidth=5)

Enter.bind("<Return>",lambda event:CalEnterBut(D.get())) #Call to CalEnterBut Function With Present Data of Entry Box When Enter Key is Pressed (Here We Achive the Enter Key Pressed Calling and Button Calling With Same Function But By USing Lambda ) ---> Here lmabda Automatically Add Event While Calling the CalEnter Function

But1=Button(Window,text=" Submit ",font=("Leto",25,"bold"),command=lambda: CalEnterBut(D.get()),bg="brown",fg="white",relief="groove",bd=8,highlightthickness=7,activebackground="black",activeforeground="cornsilk") #Automatically Call the CalEnterBut Function By Passing that Data WHich is Presnt in Entry Widget Using the Lambda Function

Enter.pack(pady=25,padx=40)
But1.pack(padx=10,pady=40)

Lab=Label(Window,text="To Submit Your Equation Kindly Press Enter or Click On Submit Button",font=("Arial",15,"bold"),bg="grey",fg="white",bd=3.5).pack()

#Set Image On Window (By Adding Frame)
Fra=Frame(Window)
Img=PhotoImage(file="C:\\Users\\HP\\Python Additional Projects (For Practice)\\Python Calculator Project\\Calc2.png")

L=Label(Fra,image=Img,bg="moccasin")

Fra.pack(pady=7)
L.pack()

#Menubar
MBar=Menu(Window) #Main MenuBar Creation

Help=Menu(MBar,bg="whitesmoke",fg="black",font=("Arial",11,"italic")) #Add Menu on Menubar which called Help
MBar.add_cascade(menu=Help,label="Help") #Dispaly SubMenu

Help.add_separator()
Help.add_command(label=" Instructions for Calculator ",command=Instructions)
Help.add_separator()

Quit=Menu(MBar,bg="whitesmoke",fg="black",font=("Arial",11,'italic'))
MBar.add_cascade(menu=Quit,label="Quit")

Quit.add_command(label=" Yes ",command=QuitYes)
Quit.add_separator()
Quit.add_command(label=" No ",command=None)
Quit.add_separator()

Window.config(menu=MBar)
Window.mainloop()
```
